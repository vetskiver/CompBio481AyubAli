```python
import pandas as pd

# List of file paths
file_paths = [
    './dataset_files/GSE140829.csv',
    './dataset_files/GSE140830.csv',
    './dataset_files/GSE140831.csv'
]

# Dictionary to hold the DataFrames
dataframes = {}

for file_path in file_paths:
    # Read the CSV file without headers
    df = pd.read_csv(file_path, header=None)

    # Drop the first and last rows
    df = df.drop(df.index[0])  # Drop first row
    df = df.drop(df.index[-1])  # Drop last row

    # Set the new first row as the header
    new_header = df.iloc[0]  # Capture the first row for the header
    df = df[1:]  # Take the data below the first row for the new DataFrame
    df.columns = new_header  # Set the captured first row as the column names

    # Reset the index of the DataFrame
    df.reset_index(drop=True, inplace=True)

    # Extract a simple identifier from the file path and use it as a key in the dictionary
    identifier = file_path.split('/')[-1].replace('.csv', '')
    
    # Store the DataFrame in the dictionary
    dataframes[identifier] = df

# Access each DataFrame using its identifier as the key
# For example, to access the DataFrame for 'GSE140829.csv'
df_GSE140829 = dataframes['GSE140829']

# Similarly, for 'GSE140830.csv' and 'GSE140831.csv'
df_GSE140830 = dataframes['GSE140830']
df_GSE140831 = dataframes['GSE140831']
```

```python
merged_df = pd.merge(dataframes['GSE140829'], dataframes['GSE140830'], on=new_header[0], how='inner')
merged_df = pd.merge(merged_df, dataframes['GSE140831'], on=new_header[0], how='inner')
merged_df.columns = [col.rstrip('_x').rstrip('_y') for col in merged_df.columns]
merged_df
```
